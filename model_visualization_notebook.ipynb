## 6. Additional Analysis Tools

### 6.1 Parameter Distribution Analysis

```python
# Analyze parameter distribution
def analyze_parameter_distribution(model, model_name):
    """Analyze the distribution of parameters in the model."""
    
    weights = []
    layer_names = []
    layer_types = []
    
    for name, param in model.named_parameters():
        if param.requires_grad:
            weights.extend(param.data.flatten().cpu().numpy())
            layer_names.extend([name] * param.numel())
            layer_types.extend([name.split('.')[-1]] * param.numel())
    
    # Create distribution plot
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Parameter value distribution
    axes[0].hist(weights, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
    axes[0].set_xlabel('Parameter Value')
    axes[0].set_ylabel('Frequency')
    axes[0].set_title(f'{model_name} - Parameter Value Distribution')
    axes[0].grid(True, alpha=0.3)
    
    # Parameter statistics by layer type
    unique_types = list(set(layer_types))
    type_stats = []
    
    for layer_type in unique_types:
        type_weights = [w for w, t in zip(weights, layer_types) if t == layer_type]
        type_stats.append({
            'Layer Type': layer_type,
            'Mean': np.mean(type_weights),
            'Std': np.std(type_weights),
            'Count': len(type_weights)
        })
    
    # Plot parameter count by type
    counts = [stat['Count'] for stat in type_stats]
    types = [stat['Layer Type'] for stat in type_stats]
    
    axes[1].bar(range(len(types)), counts, color='lightcoral', alpha=0.8, edgecolor='black')
    axes[1].set_xlabel('Layer Type')
    axes[1].set_ylabel('Parameter Count')
    axes[1].set_title(f'{model_name} - Parameters by Layer Type')
    axes[1].set_xticks(range(len(types)))
    axes[1].set_xticklabels(types, rotation=45, ha='right')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'thesis_figures/{model_name.lower()}_parameter_analysis.pdf', dpi=300, bbox_inches='tight')
    plt.show()
    
    return type_stats

# Analyze both models
print("Analyzing parameter distributions...")
resnet_param_stats = analyze_parameter_distribution(resnet_model, 'ResNet')
efficientnet_param_stats = analyze_parameter_distribution(efficientnet_model, 'EfficientNet')
```

### 6.2 Computational Complexity Analysis

```python
# Computational complexity analysis
def compute_model_complexity(model, input_shape=(3, 224, 224)):
    """Compute FLOPs and memory usage for the model."""
    
    # This is a simplified FLOP calculation
    # For more accurate results, consider using libraries like ptflops or fvcore
    
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    
    # Estimate model size in MB
    model_size_mb = (total_params * 4) / (1024 ** 2)  # Assuming float32
    
    # Estimate memory usage during training (rough approximation)
    # Forward pass + backward pass + optimizer states
    training_memory_mb = model_size_mb * 4  # Rough estimate
    
    return {
        'Total Parameters': total_params,
        'Trainable Parameters': trainable_params,
        'Model Size (MB)': model_size_mb,
        'Estimated Training Memory (MB)': training_memory_mb
    }

# Analyze complexity for both models
resnet_complexity = compute_model_complexity(resnet_model)
efficientnet_complexity = compute_model_complexity(efficientnet_model)

print("Model Complexity Analysis:")
print("\nResNet Model:")
for key, value in resnet_complexity.items():
    if 'Parameters' in key:
        print(f"  {key}: {value:,}")
    else:
        print(f"  {key}: {value:.2f}")

print("\nEfficientNet Model:")
for key, value in efficientnet_complexity.items():
    if 'Parameters' in key:
        print(f"  {key}: {value:,}")
    else:
        print(f"  {key}: {value:.2f}")

# Create comparison chart
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

models = ['ResNet-50 Based', 'EfficientNet-B0 Based']
complexities = [resnet_complexity, efficientnet_complexity]

metrics = ['Total Parameters', 'Model Size (MB)', 'Estimated Training Memory (MB)']
x = np.arange(len(models))
width = 0.25

for i, metric in enumerate(metrics):
    values = [comp[metric] for comp in complexities]
    if metric == 'Total Parameters':
        values = [v / 1e6 for v in values]  # Convert to millions
        metric_label = 'Parameters (Millions)'
    else:
        metric_label = metric
    
    ax.bar(x + i * width, values, width, label=metric_label, alpha=0.8)

ax.set_xlabel('Model')
ax.set_ylabel('Value')
ax.set_title('Model Complexity Comparison')
ax.set_xticks(x + width)
ax.set_xticklabels(models)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('thesis_figures/model_complexity_comparison.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
